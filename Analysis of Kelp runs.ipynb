```python
import pandas as pd
#import pingouin as pg
import numpy as np
from datascience import *

import statsmodels.api as sm
import statsmodels.formula.api as smf

# These lines do some fancy plotting magic.
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import warnings
warnings.simplefilter('ignore', FutureWarning)
from datetime import datetime
```

```python
# Change Column Names
rename_map = {
    '[step]': 'timestep',
    'count turtles with [color = green]': 'kelps',
    'count turtles with [shape = "circle"]': 'urchins'
}

#Load data as pandas dataframes with new column names
def load_with_stars(path, stars_label):
    return (pd.read_csv(path)
              .rename(columns=rename_map)
              .assign(stars=stars_label))

star12_pd = load_with_stars("kelp_Starfish_12-spreadsheet.csv", "twelve")
star8_pd  = load_with_stars("kelp_Starfish_8-spreadsheet.csv",  "eight")
star4_pd  = load_with_stars("kelp_Starfish_4-spreadsheet.csv",  "four")

# Combine into a single dataframe
star_pd = pd.concat([star12_pd, star8_pd, star4_pd], ignore_index=True)

# Sanity check
print(star_pd.columns)
print(star_pd.head())
```

```python
#Linear Regression Model
model = smf.ols('kelps ~ timestep * stars', data=star_pd).fit()

print(model.summary())
```
